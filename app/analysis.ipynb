```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dataset/skill_counts.csv')
```

```python
filtered_df = df.loc[df['Skill'].isin(['phd', 'masters', 'graduate degree', 'bachelor'])]
```

```python
grouped = filtered_df['Count'].groupby(filtered_df['Skill']).sum()

plt.figure(figsize=(8,8))
plt.pie(grouped.values, labels=grouped.index, autopct='%1.1f%%', startangle=140)

centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

plt.title('Degree Distribution')
plt.axis('equal')
plt.show()
```
